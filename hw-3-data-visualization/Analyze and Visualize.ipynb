### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
a = np.random.rand(7, 7)
mean = a.mean()
std = a.std()
norm = (a - mean) / std

print(f"a: {a}")
print(f"mean: {mean}")
print(f"std: {std}")
print(f"norm: {norm}")
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
total_bill_fem = tips_copy.total_bill.where(tips_copy['sex'] == 'Female')
tip_fem = tips_copy.tip.where(tips_copy['sex'] == 'Female')
smoker_fem = tips_copy.smoker.where((tips_copy['sex'] == 'Female'))

total_bill_male = tips_copy.total_bill.where(tips_copy['sex'] == 'Male')
tip_male = tips_copy.tip.where(tips_copy['sex'] == 'Male')
smoker_male = tips_copy.smoker.where((tips_copy['sex'] == 'Male'))

tbf_np = np.array(total_bill_fem.tolist())
tpf_np = np.array(tip_fem.tolist())
smf_np = np.array(smoker_fem.tolist())

tbm_np = np.array(total_bill_male.tolist())
tpm_np = np.array(tip_male.tolist())
smm_np = np.array(smoker_male.tolist())

figure, ax = plt.subplots(2)
ax[0].scatter(x=tbf_np, y=tpf_np, c=smf_np, cmap='bwr')
ax[0].set_title('Female')
ax[0].set_ylabel('Tip')
ax[1].scatter(x=tbm_np, y=tpm_np, c=smm_np, cmap='bwr')
ax[1].set_title('Male')
ax[1].set_xlabel('Total bill')
ax[1].set_ylabel('Tip')

ax[0].set_xlim(0, 55)
ax[0].set_ylim(0, 11)

ax[1].set_xlim(0, 55)
ax[1].set_ylim(0, 11)

plt.plot()
```

## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

### 1. В выходные чек срденем на 20.14% больше, чем в будние дни

```python
days = list(tips['day'].unique())
total_bill_mean_by_day = {}
for day in days:
    print(f"For {day} the mean total bill is: {tips.total_bill.where(tips.day == day).mean()}")
    total_bill_mean_by_day[day] = tips.total_bill.where(tips.day == day).mean()
# print(total_bill_mean_by_day)


tips.total_bill.hist(by=tips['day'])
plt.show()

weekend_mean = (total_bill_mean_by_day['Sun'] + total_bill_mean_by_day['Sat']) / 2
weekday_mean = (total_bill_mean_by_day['Thur'] + total_bill_mean_by_day['Fri']) / 2

weekend_mean / weekday_mean * 100 - 100
```

### 2. Для чеков выше медианного, среднее значение чаевых, оставляемых мужчинами, на 2.52% больше, чем оставляемых женщинами

```python
median_bill = tips.total_bill.median()
greater_than_median_bill = tips.where(tips.total_bill > median_bill)
dictionary = {}
sexes = list(tips['sex'].unique())
for sex in sexes:
    val = greater_than_median_bill.tip.where(greater_than_median_bill['sex'] == sex).mean()
    print(f"For {sex} median bill is {val}")
    dictionary[sex] = val

dictionary['Male'] / dictionary['Female'] * 100 - 100
```
